```python
lok_sabha_dated = []

for timestamp in list(lok_sabha["timestamp"]):
    lok_sabha_dated.append(timestamp.date())
```

```python
lok_sabha_dated.sort()
```

```python
lok_sabha_dated
```

```python
len(lok_sabha_dated)
```

```python
import collections 
  
def CountFrequency(arr): 
    return collections.Counter(arr)   

freq = CountFrequency(lok_sabha_dated)
```

```python
freq
```

```python
pdf = pd.DataFrame.from_dict(freq, orient='index').reset_index()
```

```python
pdf
```

```python
pdf.to_csv(r'pdfelections.csv')
```
