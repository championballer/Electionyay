```python
filtered = data_hashtags[data_hashtags["text"].str.contains("#")]
```

```python
filtered = list(filtered["text"].unique())
```

```python
filtered
```

```python
corpus = ""
```

```python
for text in filtered:
    corpus+=text
```

```python
tokens = corpus.split(" ")
```

```python
hashtags = []

for word in tokens:
    if word.startswith("#"):
        hashtags.append(word)
```

```python
hashtags
```

```python
import collections

def CountFrequency(tokens):
    return collections.Counter(tokens)

freq = CountFrequency(hashtags)
```

```python
freq.most_common()
```

```python
for key, cnts in list(freq.items()):   # list is important here
    if cnts < 2:
        del freq[key]
```

```python
data_hashtags = data[data['text'].str.contains("#")]
```

```python
data_hashtags
```

```python
data_hashtags.drop(['fullname', 'html','id','likes','replies','retweets','url','user'], axis=1,inplace=True)
```

```python
data_hashtags.sort_values(by="timestamp")
```

```python
text_hashtag = list(data_hashtags["text"])
```

```python
hashs = []
```

```python
import re

for index, row in data_hashtags.iterrows():
    ls = list(re.findall("[#]\w*", row['text']))
    hashs.append(ls)
```

```python
hashs
```

```python
data_hashtags2 = data_hashtags.copy()
```

```python
data_hashtags2["hashes"] = hashs
```

```python
data_hashtags2.drop(["text"],axis=1,inplace=True)
```

```python
data_hashtags2.sort_values(by="timestamp",inplace=True)
```

```python
data_hashtags2
```

```python
hashes_date_freq = {}

for index,row in data_hashtags2.iterrows():
    date = row["timestamp"].date()
    for hashe in row['hashes']:
        if hashe in hashes_date_freq.keys():
            if date in hashes_date_freq[hashe].keys():
                hashes_date_freq[hashe][date]+=1
            else:
                hashes_date_freq[hashe][date]=1
        else:
            hashes_date_freq[hashe]={}
            hashes_date_freq[hashe][date]=1
```

```python
hashes_date_freq
```

```python
from datetime import timedelta, date
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
start_date = date(2019, 1, 1)
end_date = date(2019, 5, 20)
for single_date in daterange(start_date, end_date):
    print(single_date.strftime("%Y-%m-%d"))
```
